```python
a) Erzeugen Sie die Spalte director_actor_facebook_likes als Summe aus
director_facebook_likes, actor_1_facebook_likes, actor_2_facebook_likes, actor_3_facebook_likes.
b) Erklären Sie, warum NaN-Werte entstehen.
c) Ermitteln Sie, wie viele Einträge betroffen sind.
d) Beheben Sie das Problem, indem Sie die betroffenen Felder auf 0 setzen und die Summe neu berechnen.
```

```python
like_cols = [
    "director_facebook_likes",
    "actor_1_facebook_likes",
    "actor_2_facebook_likes",
    "actor_3_facebook_likes",
]
present = [c for c in like_cols if c in movies.columns]

# a) Naive Summe (NaN propagiert, wenn Werte fehlen)
movies["director_actor_facebook_likes"] = movies[present].sum(axis=1, min_count=len(present))

# b) Ursache: Mindestens eine der beteiligten Spalten enthält NaN -> Zeilensumme wird NaN
# c) Anzahl betroffener Einträge:
movies["director_actor_facebook_likes"].isna().sum()

# d) Fehlende Werte der Basis-Spalten als 0 interpretieren und neu summieren
for c in present:
    movies[c] = pd.to_numeric(movies[c], errors="coerce").fillna(0)

movies["director_actor_facebook_likes"] = movies[present].sum(axis=1)
movies["director_actor_facebook_likes"].isna().sum()  # sollte nun 0 sein
```

```python
Ist ein Film mit Budget = 22 Mio. „kostspielig“? Wie könnte man das herausfinden?
```

```python
import numpy as np

budget_col = "budget" if "budget" in movies.columns else [c for c in movies.columns if "budget" in c.lower()][0]
budgets = pd.to_numeric(movies[budget_col], errors="coerce").dropna()

summary = budgets.describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95])
value = 22_000_000

# Empirisches Perzentil von 22 Mio.
percentile = (budgets <= value).mean() * 100

# Einfache Regel: > 75. Perzentil ⇒ „kostspielig“
is_costly = value > summary["75%"]

print(summary)
print(f"\n22 Mio. ≙ ca. {percentile:.1f}. Perzentil")
print("Einschätzung:", "kostspielig" if is_costly else "nicht kostspielig")
```

```python
SMS-Spam-Dataset (5574 Nachrichten, „spam“ oder „ham“).
a) Laden Sie den Datensatz (Plain text).
b) Fügen Sie eine Spalte hinzu, die die Gesamtlänge der Nachricht angibt.
```

```python
import pandas as pd

# robustes Laden (Encoding-Fallback)
try:
    spam = pd.read_csv("spam.csv", encoding="utf-8", low_memory=False)
except UnicodeDecodeError:
    spam = pd.read_csv("spam.csv", encoding="latin-1", low_memory=False)

# Textspalte robust finden (UCI-Format nutzt oft 'v2' für den Nachrichtentext)
text_col = None
for candidate in ["message", "v2", "text", "sms"]:
    if candidate in spam.columns:
        text_col = candidate
        break
if text_col is None:
    text_col = spam.columns[1] if len(spam.columns) >= 2 else spam.columns[0]

# b) Länge der Nachricht
spam["text_length"] = spam[text_col].astype(str).str.len()

# Stichprobe anzeigen
spam[[text_col, "text_length"]].head(10)
```
